```python
import pandas as pd
```

DataLoad

```python
df=pd.read_csv('train.csv')
```

```python
df.head()
```

```python
print(df.shape)
```

```python
print(df.info())
```

Null-Values-Check

```python
print(df.isnull().sum())
```

Correct Null Value

```python
df['Postal Code']=df['Postal Code'].fillna(0)
```

Group by


```python
category_sales=df.groupby('Category')['Sales'].sum()
print(category_sales)
```

Sort Data

```python
category_sales=df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
print(category_sales)
```

Total sales in Region

```python
region_sales=df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
print(region_sales)
```

Region + Category Group By

```python
region_category=df.groupby(['Region','Category'])['Sales'].sum().sort_values(ascending=False)
print(region_category)
```

Clean Data save 

```python
df.to_csv('clean_sales_data.csv', index=False)
print("File saved successfullu!")
```

```python
orders=df[['Order ID','Customer Name','Sales']].head(10)
regions=df[['Order ID','Region']].head(10)
```

Use apply()

```python
def sales_category(sales):
    if sales>1000:
        return 'High'
    elif sales>500:
        return 'Medium'
    else:
        return 'Low'
    
df['Sales Category']=df['Sales'].apply(sales_category)
print(df[['Sales','Sales Category']].head(10))
```

Save into Parquet file

```python
df.to_parquet('clean_sales_data.parquet',index=False)
print("Parquet file saved")
```
